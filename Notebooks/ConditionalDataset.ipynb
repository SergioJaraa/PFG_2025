```python
input_root = "D:/PFG/PFG_2025/Data/Training"
output_root = "D:/PFG/PFG_2025/Data/Training_resized"

# Clases de tumor (subcarpetas)
clases = ["glioma", "meningioma", "notumor", "pituitary"]

# Dimensión deseada
target_size = (256, 256)

for clase in clases:
    input_dir = os.path.join(input_root, clase)
    output_dir = os.path.join(output_root, clase)
    os.makedirs(output_dir, exist_ok=True)

    for filename in os.listdir(input_dir):
        if filename.lower().endswith(".jpg"):
            input_path = os.path.join(input_dir, filename)
            output_path = os.path.join(output_dir, filename)

            try:
                img = Image.open(input_path).convert("RGB")
                img_resized = img.resize(target_size, Image.LANCZOS)
                img_resized.save(output_path)
                print(f"[✓] {clase}/{filename} redimensionada.")
            except Exception as e:
                print(f"[ERROR] {filename}: {e}")
```

```python
resolutions = set()
folder = r"D:\PFG\PFG_2025\Data\Training_resized"
for fname in os.listdir(folder):
    if fname.lower().endswith(('.jpg', '.png', '.jpeg')):
        img = Image.open(os.path.join(folder, fname))
        resolutions.add(img.size)

print(resolutions)
```

```python
# Organization in subdirectories

resized_input = r"D:\PFG\PFG_2025\Data\Training_conditional_resized"
final_output = r"D:\PFG\PFG_2025\Data\Training_conditional_subfolders"

clases = ['glioma', 'meningioma', 'pituitary', 'notumor']
for clase in clases:
    os.makedirs(os.path.join(final_output, clase), exist_ok=True)

for fname in os.listdir(resized_input):
    for clase in clases:
        if fname.lower().startswith(clase):
            src = os.path.join(resized_input, fname)
            dst = os.path.join(final_output, clase, fname)
            shutil.copy2(src, dst)
            break  # evita múltiples coincidencias

print("Finished.")
```
